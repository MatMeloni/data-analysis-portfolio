```python
import pandas as pd

# Criando o conjunto de dados fict칤cio
data = {
    'CustomerID': [1, 2, 3, 4],
    'Name': ['Alice Johnson', 'Bob Smith', 'Charlie Brown', 'David Wilson'],
    'City': ['New York-NY', 'Los Angeles-CA', 'New York-NY', 'Chicago-IL'],
    'PurchaseAmount': [200, 150, 300, 250],
    'Gender': ['Female', 'Male', 'Male', 'Male'],
    'Membership': ['Premium', 'Standard', 'Standard', 'Premium']
}

df = pd.DataFrame(data)

# Exibir o DataFrame original
print("游늷 DataFrame original:")
print(df)
```

```python
# Transformando vari치veis categ칩ricas em dummies
df = pd.get_dummies(df, columns=['Gender', 'Membership'], drop_first=True)

print("游늷 DataFrame com vari치veis dummies:")
print(df)
```

```python
# Adicionando uma coluna 'Country'
df['Country'] = ['USA', 'USA', 'USA', 'USA']

# Criando vari치veis dummies para 'Country'
df = pd.get_dummies(df, columns=['Country'], drop_first=True)

print("游늷 DataFrame ap칩s adicionar e transformar 'Country':")
print(df)
```

```python
# Separando 'City' em 'CityName' e 'State'
df[['CityName', 'State']] = df['City'].str.split('-', expand=True)
df.drop(columns=['City'], inplace=True)

print("游늷 DataFrame ap칩s separar 'City' em 'CityName' e 'State':")
print(df)
```

```python
# Removendo duplicatas e verificando valores ausentes
df.drop_duplicates(inplace=True)
df.dropna(inplace=True)

print("游늷 DataFrame limpo e finalizado:")
print(df)

```

```python
import requests

# Fun칞칚o para obter dados de uma API
def fetch_data(url):
    response = requests.get(url)
    response.raise_for_status()  # Verifica se ocorreu um erro
    return response.json()

# URLs da API
users_url = 'https://jsonplaceholder.typicode.com/users'
posts_url = 'https://jsonplaceholder.typicode.com/posts'

# Obter dados dos usu치rios e posts
users_df = pd.DataFrame(fetch_data(users_url))
posts_df = pd.DataFrame(fetch_data(posts_url))

# Exibir os primeiros registros
print("\n游늷 DataFrame dos Usu치rios:")
print(users_df.head())

print("\n游늷 DataFrame dos Posts:")
print(posts_df.head())
```

```python
import matplotlib.pyplot as plt

# Contagem de posts por usu치rio
user_posts_count = posts_df['userId'].value_counts()

# Criando o gr치fico por meio de matplotlib
user_posts_count.plot(kind='bar', title='N칰mero de Posts por Usu치rio')
plt.xlabel('ID do Usu치rio')
plt.ylabel('N칰mero de Posts')
plt.show()
```
