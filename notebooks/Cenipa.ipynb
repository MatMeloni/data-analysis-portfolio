```python
# A) Importação das bibliotecas essenciais para análise de dados e visualização

import pandas as pd              # Manipulação de dados
import matplotlib.pyplot as plt  # Visualização de gráficos
import seaborn as sns            # Visualização avançada de gráficos

# Configurações visuais
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (12, 6)
```

```python
# B) Carregando o conjunto de dados da CENIPA

# Observação: o arquivo foi salvo com separador ";" e encoding "latin1"
df = pd.read_csv('/content/drive/MyDrive/Cenipa.csv', sep=';', encoding='latin1')

# Visualizando confirmação do carregamento
print("Conjunto de dados carregado com sucesso!")
```

```python
# C)
# Exibe as 5 primeiras linhas
print("Primeiras linhas do conjunto de dados:")
display(df.head())

# Exibe as 5 últimas linhas
print("Últimas linhas do conjunto de dados:")
display(df.tail())
```

```python
#4) Análise das informações do conjunto de dados
#nformações gerais sobre o DataFrame
print("Informações do DataFrame:")
df.info()

# Estatísticas descritivas para todas as colunas (inclusive texto)
print("\nEstatísticas descritivas:")
display(df.describe(include='all'))
```

```python
 # 5) Verificação de dados ausentes

 # Total de valores ausentes por coluna
missing_data = df.isnull().sum()

# Porcentagem de valores ausentes
missing_percent = (missing_data / len(df)) * 100

# Tabela combinada com total e percentual
missing_df = pd.DataFrame({
    'Total Faltante': missing_data,
    '% Faltante': missing_percent
})

# Exibir apenas colunas com valores faltantes
missing_df = missing_df[missing_df['Total Faltante'] > 0]

# Ordenar da maior para a menor
missing_df = missing_df.sort_values(by='% Faltante', ascending=False)

# Exibir resultado
print("Colunas com dados ausentes:")
display(missing_df)
```

```python
# 6) Distribuição do tipo de acidente (com filtro e visualização otimizada)

# Verifica se a coluna existe no DataFrame
if 'Tipo de Ocorrencia' in df.columns:
    # Contagem de cada tipo de ocorrência
    ocorrencia_counts = df['Tipo de Ocorrencia'].value_counts()

    # Filtra os tipos com mais de 30 ocorrências
    ocorrencia_filtrada = ocorrencia_counts[ocorrencia_counts > 30]

    # Gráfico de barras horizontal
    plt.figure(figsize=(10, 8))
    sns.barplot(y=ocorrencia_filtrada.index, x=ocorrencia_filtrada.values, palette='Blues_r')
    plt.title('Tipos de Ocorrência com mais de 30 Registros')
    plt.xlabel('Quantidade')
    plt.ylabel('Tipo de Ocorrência')
    plt.tight_layout()
    plt.show()
else:
    print("A coluna 'Tipo de Ocorrencia' não está presente no DataFrame.")


```

```python
# 7) Análise das causas dos acidentes
# Verifica se a coluna existe
if 'Fator do Acidente' in df.columns:
    # Conta quantas vezes cada fator aparece
    fator_counts = df['Fator do Acidente'].value_counts()

    # Filtra apenas fatores com mais de 10 ocorrências
    fator_filtrado = fator_counts[fator_counts > 10]

    # Plota o gráfico somente com os mais relevantes
    plt.figure(figsize=(10, 6))
    sns.barplot(x=fator_filtrado.values, y=fator_filtrado.index, orient='h', color='steelblue')
    plt.title('Fatores Contribuintes com mais de 10 Ocorrências')
    plt.xlabel('Quantidade')
    plt.ylabel('Fator do Acidente')
    plt.tight_layout()
    plt.show()
else:
    print("A coluna 'Fator do Acidente' não está presente no DataFrame.")
```

```python
# 8) Análise temporal dos acidentes

# Verifica se a coluna existe no DataFrame
if 'Data e Hora da Ocorrencia' in df.columns:
    # Converte a coluna para o formato de data
    df['Data e Hora da Ocorrencia'] = pd.to_datetime(df['Data e Hora da Ocorrencia'], errors='coerce')

    # Cria uma nova coluna com o ano
    df['Ano'] = df['Data e Hora da Ocorrencia'].dt.year

    # Agrupa por ano e conta quantas ocorrências
    acidentes_por_ano = df['Ano'].value_counts().sort_index()

    # Plota o gráfico
    plt.figure(figsize=(12, 6))
    sns.barplot(x=acidentes_por_ano.index, y=acidentes_por_ano.values, palette='viridis')
    plt.title('Quantidade de Ocorrências por Ano')
    plt.xlabel('Ano')
    plt.ylabel('Quantidade de Ocorrências')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
else:
    print("A coluna 'Data e Hora da Ocorrencia' não está presente no DataFrame.")
```

```python
# 9) Análise de acidentes por região (UF)

# Verifica se a coluna existe no DataFrame
if 'UF da Ocorrencia' in df.columns:
    # Contagem de ocorrências por estado (UF)
    uf_counts = df['UF da Ocorrencia'].value_counts()

    # Filtra UFs com mais de 10 ocorrências para deixar o gráfico limpo
    uf_filtradas = uf_counts[uf_counts > 10]

    # Gráfico de barras horizontal
    plt.figure(figsize=(10, 8))
    sns.barplot(x=uf_filtradas.values, y=uf_filtradas.index, palette='mako')
    plt.title('Estados com mais de 10 Ocorrências Registradas')
    plt.xlabel('Quantidade de Ocorrências')
    plt.ylabel('UF (Estado)')
    plt.tight_layout()
    plt.show()
else:
    print("A coluna 'UF da Ocorrencia' não está presente no DataFrame.")
```

```python
# 10) Correlação entre variáveis numéricas

# Seleciona apenas colunas numéricas
numeric_df = df.select_dtypes(include=['int64', 'float64'])

# Verifica se há variáveis numéricas suficientes para análise
if not numeric_df.empty and numeric_df.shape[1] >= 2:
    # Calcula a matriz de correlação
    correlation = numeric_df.corr()

    # Plota o mapa de calor
    plt.figure(figsize=(10, 8))
    sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt=".2f", square=True)
    plt.title('Correlação entre Variáveis Numéricas')
    plt.tight_layout()
    plt.show()
else:
    print("Não há variáveis numéricas suficientes para gerar uma matriz de correlação.")
```
