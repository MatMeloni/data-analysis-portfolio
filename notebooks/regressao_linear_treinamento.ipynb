```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Dados da atividade
horas_treinamento = np.array([5.2, 5.1, 4.9, 4.6, 4.7, 4.8, 4.6, 4.9]).reshape(-1, 1)
tempo_trabalho = np.array([13, 15, 16, 20, 19, 17, 21, 16])

# Criando o modelo
modelo = LinearRegression()
modelo.fit(horas_treinamento, tempo_trabalho)

# Previs√£o
tempo_predito = modelo.predict(horas_treinamento)

# Par√¢metros do modelo
coef_angular = modelo.coef_[0]
intercepto = modelo.intercept_
mse = mean_squared_error(tempo_trabalho, tempo_predito)

# Resultados
print(f'Equa√ß√£o da reta: y = {coef_angular:.2f}x + {intercepto:.2f}')
print(f'MSE: {mse:.2f}')
```

```python

# Gr√°fico de dispers√£o com linha de regress√£o
plt.figure(figsize=(10, 5))
plt.scatter(horas_treinamento, tempo_trabalho, color='blue', label='Dados reais')
plt.plot(horas_treinamento, tempo_predito, color='red', label='Linha de regress√£o')
plt.xlabel("Horas de Treinamento")
plt.ylabel("Tempo para completar o trabalho")
plt.title("Atividade Pr√°tica: Introdu√ß√£o √† Modelagem Estat√≠stica e An√°lise de Regress√£o Simples")
plt.legend()
plt.grid(True)
plt.show()
```

O coeficiente angular **(-11,91)** indica que quanto mais treinamento, menor o tempo de execu√ß√£o.

O intercepto **(74,90)** indica o tempo te√≥rico para algu√©m sem treinamento.

O Erro Quadr√°tico M√©dio (MSE) foi de **0,33**, o que representa bom ajuste do modelo para esses dados.
