```python
import pandas as pd
from google.colab import drive

drive.mount('/content/drive', force_remount=True)
file_path = '/content/drive/MyDrive/Separa/Analise de Dados/Top 1000 technology companies.csv'
df = pd.read_csv(file_path)
display(df.head())
```

```python
def precisa_corrigir(row):
  return pd.isna(row['Company'])

def remover_virgulas_dentro_de_aspas(texto):
  nova_string = ""
  dentro_de_aspas = False

  for char in texto:
    if char == '"':
      dentro_de_aspas = not dentro_de_aspas
      continue
    if char == ',' and dentro_de_aspas:
      continue
    nova_string += char

  return nova_string


def remover_aspas(texto):
  nova_string = ""

  for char in texto:
    if char == '"':
      continue
    nova_string += char
  return nova_string

def encontrar_linha_para_corrigir(df):

  for index, row in df.iterrows():
    if precisa_corrigir(row):
      texto = row['Ranking']
      texto = remover_virgulas_dentro_de_aspas(texto)
      texto = remover_aspas(texto)

      dados = texto.split(',')
      dados = [dado.strip() for dado in dados]
      df.at[index, 'Ranking'] = dados[0]
      df.at[index, 'Company'] = dados[1]
      df.at[index, 'MarketCap'] = dados[2]
      df.at[index, 'Stock'] = dados[3]
      df.at[index, 'Country'] = dados[4]
      df.at[index, 'Sector'] = dados[5]
      df.at[index, 'Industry'] = dados[6]

encontrar_linha_para_corrigir(df)
df.head(10)
```

```python
caracteres_especiais = "!@#$%^&*()[]{};:,/<>?\|`~-=+"


def retira_caracteres_especiais(texto):
  texto_novo = ""
  for char in texto:
    if char not in caracteres_especiais:
      texto_novo += char

  return texto_novo
def formatar_texto_colunas(texto):
  if isinstance(texto, str):
    texto = texto.lower()
    texto = texto.replace(' ', '_')
    texto = retira_caracteres_especiais(texto)
  return texto

def formatar_texto_marketcap(texto):
  if isinstance(texto, str):
    texto = texto.lower()
    texto = texto.replace(' ', '')
    texto = retira_caracteres_especiais(texto)
  return texto


def formata_textos_data_frame(df):
  for coluna in df.columns:
    if coluna != 'MarketCap':
      df[coluna] = df[coluna].apply(formatar_texto_colunas)
    else:
      df[coluna] = df[coluna].apply(formatar_texto_marketcap)

formata_textos_data_frame(df)
df.head(1000)
```

```python
valor_total_companias = 0


def converter_para_milhoes(valor):
    if 't' in valor:
        return float(valor.replace('t', '')) * 1_000_000
    elif 'b' in valor:
        return float(valor.replace('b', '')) * 1_000
    elif 'm' in valor:
        return float(valor.replace('m', ''))
    return 0


for linha_marketcap in df['MarketCap']:
    valor_total_companias += converter_para_milhoes(linha_marketcap)

print(f'O valor total das companias são de: {valor_total_companias/1_000_000:.2f}t')
```
